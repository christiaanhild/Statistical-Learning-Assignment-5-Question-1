```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(z):
    return np.maximum(0,z)

def abs_relu(z):
    return np.abs(z)

def leaky_relu(z, alpha=0.1):
    return np.maximum(0,z) + alpha * np.minimum(0,z)

def tanh(z):
    return 2 * (1 / (1+np.exp(-2 * z))) - 1

def elu(z, alpha=0.1):
    return np.array([x if x>=0 else alpha*(np.exp(x)-1) for x in z])
    
def softplus(z):
    return np.log(1+np.exp(z))
```

```python
x = np.linspace(-2,2,100)
```

```python
plt.figure(figsize=(10,6))

plt.plot(x, relu(x), color="blue", label='relu')
plt.plot(x, abs_relu(x), color="orange", label='absolute relu')
plt.plot(x, leaky_relu(x), color="green", label='leaky relu (alpha=0.1)')
plt.plot(x, tanh(x), color="red", label='tanh')
plt.plot(x, elu(x), color="pink", label='elu (alpha=0.1)')
plt.plot(x, softplus(x), color="purple", label='softplus')

plt.title("Different activation functions for hidden layer.")
plt.legend()
plt.show()
```
